```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn import datasets
from sklearn.mixture import GaussianMixture

# load the dataset
iris = datasets.load_iris()
X = iris.data[:, :2]  # we'll use only the first two features for simplicity

# create a Gaussian mixture model with 3 compinents/cluster
gmm = GaussianMixture(n_components=3)

# fit the model to the data
gmm.fit(X)

# predict the cluster labels
labels = gmm.predict(X)

# visualize the cluster model
plt.scatter(X[:,0], X[:,1], c=labels, cmap='viridis')
plt.xlabel('Sepal length')
plt.ylabel('Sepal width')
plt.title('guassian mixture model on clustering on Iris dataset')
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn import datasets
from sklearn.mixture import GaussianMixture

# load the dataset
iris = datasets.load_iris()
X = iris.data[:, :2]  # we'll use only the first two features for simplicity

# create a Gaussian mixture model with 3 compinents/cluster
gmm = GaussianMixture(n_components=3)

# fit the model to the data
gmm.fit(X)

# predict the cluster labels
labels = gmm.predict(X)

# compute the log-likelihood of the data
log_likelihood = gmm.score(X)

# compute the AIC or BIC 
n_sample = X.shape
aic = gmm.aic(X) # AIC (Akaike information Criterion)
bic = gmm.bic(X) # BIC (Bayesian Information Criterion)

# print the log-likelihood 
print('log-likelihood:', log_likelihood)
print('AIC:', aic)
print('BIC:', bic)

# visualize the cluster model
plt.scatter(X[:,0], X[:,1], c=labels, cmap='viridis')
plt.xlabel('Sepal length')
plt.ylabel('Sepal width')
plt.title('guassian mixture model on clustering on Iris dataset')
plt.show()
```
