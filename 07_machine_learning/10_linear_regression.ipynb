# Linear Regression 
- **Simple Regression** : only one independent variable (e.g. sales)
- **Multiple Regression** : more than one independent variable (e.g. sales, pirce)

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error, r2_score
```

```python
# load the data
df = sns.load_dataset('tips')
df.head()
```

```python
sns.scatterplot(df, x='total_bill', y='tip')
```

```python
# split the data into X and y
X = df[['total_bill']]
# scaler = StandardScaler()
# X = scaler.fit_transform(X)
y = df['tip']
```

```python
# split the data into train and test
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)
```

```python
# call the model 
model = LinearRegression()
```

```python
# train the model 
model.fit(X_train, y_train)
```

```python
# Take out the intercept slop, make an equation
print(model.intercept_)
print(model.coef_)
print('y=', model.intercept_, '+', model.coef_, "*X")
```

```python
model.predict([[10]])
```

```python
# predict
y_pred = model.predict(X_test)
```

```python
# evaluate the model
print('MSE', mean_squared_error(y_test, y_pred))
print('r2', r2_score(y_test, y_pred))
print('RMSE', np.sqrt(mean_squared_error(y_test, y_pred)))
```

```python
# plot the model and data
plt.scatter(X_test, y_test)
plt.plot(X_test, y_pred, color = "r")
plt.show()
```
