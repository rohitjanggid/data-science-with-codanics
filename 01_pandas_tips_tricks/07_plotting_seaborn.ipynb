# Seaborn and Plotting

```python
import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

### Important thinks for plotting according to grammer of graphics

1. data(df)
2. type of plot?
3. x axis
4. y axis 
5. grouping (hue/color) based on categorical variable
6. Size of the plot
7. Collor palletes of the plot
8. Style of the plot
9. How do you interpret

```python
df = sns.load_dataset('titanic')
```

```python
df.info()
```

```python
# scatter plot
# scatter plot size of figure size
plt.figure(figsize=(8,6))
# change the scatter point color
palette = {'male': 'green', 'female' : 'red'}
sns.scatterplot(data = df, x='age', y='fare', hue='sex',size='age', palette=palette)
# label
plt.xlabel('Age')
plt.ylabel('Fare')
plt.title('Scatter plot of Age vs Fare')

# Saving a plot 300dpi png
plt.savefig('./outputs/plots/scatter_plot.png', dpi= 300)

# Saving a plot in vector file such as svg
plt.savefig('./outputs/plots/scatter_plot.svg', dpi= 300)
```

```python
# style of the plot
sns.set_theme(style='darkgrid', font_scale = 1.5, font='Times New Roman')

# scatter plot size of figure size
plt.figure(figsize=(16,9))
# change the scatter point color
palette = {'male': 'green', 'female' : 'red'}
sns.scatterplot(data = df, x='age', y='fare', hue='sex',size='age', palette=palette)
# label
plt.xlabel('Age')
plt.ylabel('Fare')
plt.title('Scatter plot of Age vs Fare')
```

Interpretation :
1. We have age of passengers at x-axis and fare at y-asix
2. The scatter plot show the relastionship between age and fare of titanic passengers.
3. We furhter grouped the passenger by their genders such as male(green points) and female(red points)
4. Two male passenger gave the highest fare in titanic, and the might be traveling first class.
5. the oldest prosen in titanic is also a male having 80 years of age as shown in the scatter plot. (as shown in figure. 1)

# Box plot

```python
df['class'].unique()
```

```python
# style of the plot
sns.set_theme(style='darkgrid', font_scale = 1, font='Times New Roman')

# box plot size of figure size
plt.Figure(figsize=(16,9))
# change the box point color
palette = {'First': 'green', 'Second' : 'red', 'Third' : 'blue'}
sns.boxplot(data = df, x='sex', y='age', hue='class', palette=palette)
# label
plt.xlabel('sex')
plt.ylabel('Fare')
plt.title('box plot of Age vs Fare')

# change legend position and title
plt.legend(loc = 'upper right',  title = 'Class Name')
```

```python
# style of the plot
sns.set_theme(style='darkgrid', font_scale = 1, font='Times New Roman')

# boxen plot size of figure size
plt.Figure(figsize=(16,9))
# change the box point color
palette = {'First': 'green', 'Second' : 'red', 'Third' : 'blue'}
sns.boxenplot(data = df, x='sex', y='age', hue='class', palette=palette)
# label
plt.xlabel('Gender')
plt.ylabel('age')
plt.title('box plot of age based on gender')

# change legend position and title
plt.legend(loc = 'upper right',  title = 'Class Name')

# change the x-tick name
plt.xticks(ticks=[0,1], labels=['Male', 'Female'], rotation=45)
```

```python
# Make scatter plot with plotly
import plotly.express as px
fig = px.scatter(df, x='age', y='fare', color='sex')
fig.show()
```

```python
# bar plot using seaborn
# style of the plot
sns.set_theme(style='darkgrid', font_scale = 1, font='Times New Roman')

# bar plot size of figure size
plt.Figure(figsize=(16,9))
# change the bar point color
palette = {'First': 'green', 'Second' : 'red', 'Third' : 'blue'}
sns.barplot(data = df, x='sex', y='age', hue='class', palette=palette, capsize=0.1)
# label
plt.xlabel('Gender')
plt.ylabel('age')
plt.title('bar plot of age based on gender')

# change legend position and title
plt.legend(loc = 'upper right',  title = 'Class Name')

# change the x-tick name
plt.xticks(ticks=[0,1], labels=['Male', 'Female'])
```

```python
# violin plot using seaborn
# style of the plot
sns.set_theme(style='darkgrid', font_scale = 1, font='Times New Roman')

# violin plot size of figure size
plt.figure(figsize=(8,6))
# change the bar point color
palette = {'First': 'green', 'Second' : 'red', 'Third' : 'blue'}
sns.violinplot(data = df, x='sex', y='age', hue='class', palette=palette)
# label
plt.xlabel('Gender')
plt.ylabel('age')
plt.title('bar plot of age based on gender')

# change legend position and title
plt.legend(loc = 'upper right',  title = 'Class Name')

# change the x-tick name
plt.xticks(ticks=[0,1], labels=['Male', 'Female'])
```
